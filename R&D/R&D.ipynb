```python
import pandas as pd
import numpy as np
import pymorphy3
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from nltk.corpus import stopwords
import string
from bs4 import BeautifulSoup


def preprocess_text(text):
    text = BeautifulSoup(text, 'html.parser').get_text(separator='\n', strip=True)
    text = ''.join([char for char in text if char not in string.punctuation])
    text = ''.join([i for i in text if not i.isdigit()])
    return text.strip().lower()


def lemmatize_text(text):
    morph = pymorphy3.MorphAnalyzer()
    words = text.split()
    lemmas = [morph.parse(word)[0].normal_form for word in words]
    return ' '.join(lemmas)


def remove_stopwords(text):
    stops = set(stopwords.words('russian'))
    words = text.split()
    filtered_words = [word for word in words if word not in stops]
    return ' '.join(filtered_words)


class TextPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self):
        super().__init__()
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.apply(preprocess_text)


class StopWordsRemover(BaseEstimator, TransformerMixin):
    def __init__(self):
        super().__init__()
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.apply(remove_stopwords)


class Lemmatizer(BaseEstimator, TransformerMixin):
    def __init__(self):
        super().__init__()
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.apply(lemmatize_text)


# –°–æ–∑–¥–∞–µ–º pipeline
pipeline = make_pipeline(
    TextPreprocessor(),
    StopWordsRemover(),
    Lemmatizer(),
    TfidfVectorizer(analyzer='word', token_pattern=r'\w+', max_features=1000)
)

# –ü—Ä–∏–º–µ–Ω—è–µ–º pipeline
X_tfidf = pipeline.fit_transform(texts_df['text'])
feature_names = pipeline.named_steps['tfidfvectorizer'].get_feature_names_out()

# –í—ã–≤–æ–¥–∏–º —Ç–æ–ø-10 —Å–ª–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–æ–∫—É–º–µ–Ω—Ç–∞
for doc_idx in range(X_tfidf.shape[0]):
    print(f"\n{'='*60}")
    print(f"–î–æ–∫—É–º–µ–Ω—Ç {doc_idx + 1}:")
    print('='*60)
    
    # –ü–æ–ª—É—á–∞–µ–º –≤–µ–∫—Ç–æ—Ä –¥–ª—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ –¥–æ–∫—É–º–µ–Ω—Ç–∞
    doc_vector = X_tfidf[doc_idx].toarray().flatten()
    
    # –°–æ—Ä—Ç–∏—Ä—É–µ–º –∏–Ω–¥–µ–∫—Å—ã –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∑–Ω–∞—á–µ–Ω–∏–π TF-IDF
    top_indices = doc_vector.argsort()[::-1][:10]
    
    # –í—ã–≤–æ–¥–∏–º —Ç–æ–ø-10 —Å–ª–æ–≤
    for idx, feature_idx in enumerate(top_indices):
        if doc_vector[feature_idx] > 0:  # –ø–æ–∫–∞–∑—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –Ω–µ–Ω—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
            print(f"{idx + 1}. –°–ª–æ–≤–æ '{feature_names[feature_idx]}' = {doc_vector[feature_idx]:.4f}")
```

```python
# –í–∞—Ä–∏–∞–Ω—Ç 2: –°—Ä–µ–¥–Ω–∏–π TF-IDF –ø–æ –≤—Å–µ–º –¥–æ–∫—É–º–µ–Ω—Ç–∞–º
print("\n" + "="*60)
print("–¢–û–ü-10 –°–õ–û–í –ü–û –°–†–ï–î–ù–ï–ú–£ TF-IDF (–ø–æ –≤—Å–µ–º –¥–æ–∫—É–º–µ–Ω—Ç–∞–º):")
print("="*60)

tfidf_mean = X_tfidf.mean(axis=0).A1
top_indices_mean = tfidf_mean.argsort()[::-1][:10]

for idx, feature_idx in enumerate(top_indices_mean):
    print(f"{idx + 1}. –°–ª–æ–≤–æ '{feature_names[feature_idx]}' = {tfidf_mean[feature_idx]:.4f}")

# –í–∞—Ä–∏–∞–Ω—Ç 3: –¢–æ–ø-—Å–ª–æ–≤–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–æ–∫—É–º–µ–Ω—Ç–∞ (—Å —É—á–µ—Ç–æ–º IDF –ø–æ –≤—Å–µ–º—É –∫–æ—Ä–ø—É—Å—É)
print("\n" + "="*60)
print("–¢–û–ü-10 –°–õ–û–í –î–õ–Ø –ö–ê–ñ–î–û–ì–û –î–û–ö–£–ú–ï–ù–¢–ê:")
print("="*60)

for doc_idx in range(X_tfidf.shape[0]):
    print(f"\n–î–æ–∫—É–º–µ–Ω—Ç {doc_idx + 1}:")
    print("-"*60)
    
    doc_vector = X_tfidf[doc_idx].toarray().flatten()
    top_indices = doc_vector.argsort()[::-1][:10]
    
    for idx, feature_idx in enumerate(top_indices):
        if doc_vector[feature_idx] > 0:
            print(f"{idx + 1}. '{feature_names[feature_idx]}' = {doc_vector[feature_idx]:.4f}")
```
